```python
import pandas as pd
```

```python
data = {'Country':['India','USA','Japan','China','India','USA'],
        'Year' : ['2010','2010','2010','2015','2015','2015'],
        'Continent':['Asia','NA','Asia','Asia','Asia','NA'],
        'Value':[20,10,40,50,60,55],
       }
```

# Concat

```python
people1 = {'id' : [1,2,3,4],
          'name': ['ab','rf','kl','ed'],
          'age': [45,25,41,30]
          }
people2 = {'id' : [11,22,33,44],
          'name': ['abc','rfg','klf','ede'],
          'age': [48,22,40,35]
          }
people3 = {'id' : [111,222,333,444],
          'name': ['rab','frf','hkl','ted'],
          'age': [49,29,43,38]
          }
```

```python
df1 = pd.DataFrame(people1)
df2 = pd.DataFrame(people2)
df3 = pd.DataFrame(people3)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
pd.concat([df1,df2,df3],axis = 1)
```

```python
people4 = {'id' : [1,2,3,4],
          'name': ['ab','rf','kl','ed'],
          'age': [45,25,41,30]
          }
people5 = {'id' : [1,2,3,4],
          'name': ['ab','rf','kl','ed'],
          'sales': [948,822,440,355],
          }
```

```python
df4=pd.DataFrame(people4)
df5=pd.DataFrame(people5)
```

```python
pd.merge(df4,df5,on=['id'])
```

```python
people6 = {'id' : [1,2,3,4],
          'name': ['ab','rf','kl','ed'],
          'age': [45,25,41,30]
          }
people7 ={
          's_name': ['ab','rf','kl','ed'],
          'ph': [955457848,8248412122,9785554440,7854565355],
          }
```

```python
df6=pd.DataFrame(people6)
df7=pd.DataFrame(people7)
```

```python
df6.join(df7)
```
